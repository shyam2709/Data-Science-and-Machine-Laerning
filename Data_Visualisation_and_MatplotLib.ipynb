```python
x=np.arange(-3,3,0.1)
y=x**2
plt.plot(x,y,'rv')                             #to change line colour to red (google matplotlib plt colour for all colour list)
plt.xticks(np.arange(-3,3,0.2),rotation=90)   #to rotate numbers on axis
plt.yticks(np.arange(0,10,0.5))
plt.title('y=x**2')                           #to give title
plt.xlabel('x_value')                         #to give label to x axis
plt.ylabel('y_value')                         #to give label to y axis
plt.show()                                    #to show at end
```

```python
x=[1994,1996,1998,2000,2002]
y=[48457,28163,41061,2348,10308]
z=[12478,942,16227,166,14832]
plt.scatter(x,y,marker='s')
#Create the plot number_of_customers and years
plt.plot(x,z)
#Add the xlabel
plt.xlabel('years')
#Add the ylabel
plt.ylabel('sales and number of customers')
#Add the legends
plt.legend(['sales','number of customers(in thousands)'])
#Add the title
plt.title('Analysis of sales and customers')
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y=x**2
plt.plot(x,y,'r|')                             #to change line colour to red (google matplotlib plt colour for all colour list)
plt.xticks(np.arange(-3,3,0.2),rotation=90)   #to rotate numbers on axis
plt.yticks(np.arange(0,10,0.5))
plt.title('y=x**2')                           #to give title
plt.xlabel('x_value')                         #to give label to x axis
plt.ylabel('y_value')                         #to give label to y axis
plt.xlim(left=0,right=3)                      # x lim is x limiter
plt.show() 
```

```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**3
plt.plot(x,y,label='square')
plt.plot(x,y1,label='cube')
plt.plot(x,y2,label='raise to 4')
plt.xticks(np.arange(-3,3,0.1),rotation=90)
plt.yticks(np.arange(0,10,0.5))
plt.title('y=x**2,y1=x**3,y2=x**3')
plt.xlabel('x_value')
plt.ylabel('y_value')
plt.legend()
plt.show()
```

```python
x=np.arange(-3,3,0.1)
y=x**2
y1=x**3
y2=x**4
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xticks(np.arange(-3,3,1),rotation=90)
plt.yticks(np.arange(-27,82,10))
plt.title('y=x**2,y1=x**3,y2=x**3')
plt.xlabel('x_value')
plt.ylabel('y_value')
plt.legend(['square','cube','raise to 4'],loc='upper center',frameon=False)    #not need to give label in each plot when legend is used,loc=location,frameon=false means frame border of frame to be removed
plt.grid()              #grid gives lines in background
plt.show()
```

Bar Graph

```python
fruits=[1,2,3]
price=[10,20,30]
plt.bar(fruits,price,width=0.4,color='g')  #for width of bars and colours
plt.title('Bar graph of fruits')
plt.xlabel('Price')
plt.ylabel('Fruit')
plt.legend(['Fruits'],loc='upper center left')
plt.show()
```

With 2 bar graphs

```python
fruits=np.array([1,2,3])
market1=np.array([10,34,24])
market2=np.array([34,63,21])
plt.bar(fruits,market1,width=0.2)
plt.bar(fruits+0.2,market2,width=0.2)
plt.legend(['Market1','Market2'])
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=np.sin(x)
y1=np.tan(x)
y2=np.cos(x)
y3=np.log(x)
fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].plot(x,y)
ax[0,0].set_title("sin")
ax[0,1].plot(x,y1)
ax[1,0].plot(x,y2)
ax[1,1].plot(x,y3)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=np.sin(x)
y1=np.cos(x)
y2=np.tan(x)
y3=np.log(x)
plt.figure(figsize=(5,5))
plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.plot(x,y1)
plt.subplot(2,3,3)
plt.plot(x,y2)
plt.subplot(2,3,4)
plt.plot(x,y3)
```
