```python
df_agg.columns=["_".join(col) for col in df_agg.columns]
```

```python
df_agg
```

```python
df_agg["years_productive"] = df_agg["year_max"]-df_agg["year_min"]
```

```python
df_agg
```

```python
df_agg["movies_per_year"]=df_agg["title_count"]/df_agg["years_productive"]
```
